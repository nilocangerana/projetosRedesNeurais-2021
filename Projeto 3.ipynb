```python
#Imports
import platform
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.mobilenet_v2 import preprocess_input
from tensorflow.keras.applications.mobilenet_v2 import MobileNetV2
```

```python
print("Versão Python utilizada: ",platform.python_version())
print("Versão Keras: ",tf.keras.__version__)
print("Versão Tensorflow: ",tf.__version__)
print("Número de GPUs Disponiveis: ", len(tf.config.list_physical_devices('GPU')))
```

## Questão 1

```python
#importar parte convolucional da MobileNetV2:
conv_model = MobileNetV2(input_shape=(128,128,3), include_top=False, weights="imagenet")
```

```python
#Congelar os parametros:
for layer in conv_model.layers[:]:
    layer.trainable = False
```

```python
#Adicionar as camadas restantes:
new_model = conv_model.output
new_model = Flatten()(new_model)
new_model = Dense(7, activation='softmax')(new_model)
```

```python
#Criar modelo:
model = Model(inputs=conv_model.input, outputs=new_model)
```

```python
#Verificar modelo completo:
model.summary()
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

## Questão 2

```python
#Image data generator
batch_size=32
num_epocas=10

datagen_train = ImageDataGenerator(
    horizontal_flip=True, 
    zoom_range=0.2, 
    shear_range=10, 
    height_shift_range=0.1, 
    width_shift_range=0.1, 
    preprocessing_function=preprocess_input
)

datagen_test = ImageDataGenerator(
    preprocessing_function=preprocess_input
)
```

```python
train_gen = datagen_train.flow_from_directory(
    'dataset\\train\\', 
    target_size=(128,128), 
    batch_size=batch_size, 
    class_mode='categorical'
)

test_gen = datagen_test.flow_from_directory(
    'dataset\\test\\', 
    target_size=(128,128), 
    batch_size=batch_size, 
    class_mode='categorical'
)
```

O test_gen foi passado no parâmetro validation_data do método fit para calcular a acurácia depois de cada época.


O resultado da acurácia para o conjunto de treino depois de cada época pode ser visto no campo accuracy de cada época.
<br>O resultado da acurácia para o conjunto de teste depois de cada época pode ser visto no campo val_accuracy de cada época.

```python
#Treinamento do modelo:
model.fit(train_gen, epochs=num_epocas, validation_data=test_gen)
```
